# Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
dfcars = sns.load_dataset('car_crashes')
dfcars.head()
```

# Dispribition plot

```python
sns.distplot(dfcars['not_distracted'])
```

```python
sns.distplot(dfcars['not_distracted'],kde=False)
```

```python
sns.distplot(dfcars['not_distracted'],kde=False,bins=60)
```

# Joint plot

```python
sns.jointplot(x='speeding',y='alcohol',data=dfcars,kind='reg');
```

```python
sns.jointplot(x='speeding',y='alcohol',data=dfcars,kind='kde');
```

```python
sns.jointplot(x='speeding',y='alcohol',data=dfcars,kind='hex');
```
