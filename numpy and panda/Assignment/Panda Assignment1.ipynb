```python
#Which was the most ordered item? and How many items were ordered?
df1.groupby('item_name').sum().sort_values('quantity',ascending=False).head(1)
```

```python
#What was the most ordered item in the choice_description column?
df1.groupby('choice_description').sum().sort_values('quantity',ascending=False).head(1)
```

```python
#Turn the item price into a float
def getPrice(a):
    return float(a[1:])

df1["Tprice"]=df1['item_price'].apply(getPrice)
df1.head()
```

```python
#print a data frame with only two columns item_name and item_price
df1[['item_name','item_price']]
```

```python
# delete the duplicates in item_name and quantity
df1[['item_name','quantity']].drop_duplicates()
```

```python
# select only the products with quantity equals to 1
df1[df1['quantity']==1]['item_name']
```

```python
# select only the item_name and item_price columns
df1[['item_name','item_price']]
```

```python
# sort item from the most to less expensive
df1.sort_values('item_price',ascending=True)
```

```python
#What was the quantity of the most expensive item ordered?
df1.sort_values('Tprice',ascending=False).head(1)['quantity']
```
